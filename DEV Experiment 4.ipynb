```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Creation (Corrected lengths)
np.random.seed(42)  # for reproducibility
df = pd.DataFrame({
    'ID': range(1, 11),
    'Age': np.random.randint(18, 65, size=10),
    'Income': np.random.randint(30000, 90000, size=10),
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male',
               'Female', 'Male', 'Female', 'Male', 'Male'],  # 10 values
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor',
                  'Master', 'Bachelor', 'PhD', 'High School', 'Master']  # 10 values
})

# 2. Basic Overview
print(" First 5 Rows:\n", df.head(), "\n")
print(" Statistical Summary:\n", df.describe(include=[np.number]), "\n")
print(" Missing Values:\n", df.isnull().sum(), "\n")
print(" Unique Genders:", df['Gender'].unique())
print(" Unique Education Levels:", df['Education'].unique(), "\n")

# 3. Subsets and Filtering
print("\n Selected Columns (Age & Income):\n", df[['Age', 'Income']].head())
print("\n Filtered (Age > 30):\n", df[df['Age'] > 30].head())
print("\n Filtered (Male & Master's Degree):\n", df[(df['Gender'] == 'Male') & (df['Education'] == 'Master')].head())

# 4. Visualizations Setup
sns.set(style="whitegrid")  # clean styling

# Histogram of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=5, kde=False, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Boxplot of Income
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['Income'], color='lightblue')
plt.title('Income Distribution')
plt.ylabel('Income')
plt.show()


# Bar plot of Gender (fixed for Seaborn 0.14+)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', hue='Gender', palette='pastel', legend=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


# Pie chart of Education
plt.figure(figsize=(6, 6))
df['Education'].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=sns.color_palette('Set2'),
    startangle=140
)
plt.title('Education Distribution')
plt.ylabel('')
plt.show()
```
