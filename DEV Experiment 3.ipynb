```python
import numpy as np 
a=np.array([[1,2,3],[3,4,5],[4,5,6]]) 
print('Our array is:')
print(a) 
print('The items in the second column are:') 
print(a[...,1]) 
print('\n') 
print('The items in the second row are:') 
print(a[1,...]) 
print('\n') 
print('The items column 1 onwards are:') 
print(a[...,1])
```

```python
import numpy as np 
import pandas as pd 

# Fixed data
data = np.array([['', 'Col1', 'Col2'], ['Row1', 1, 2], ['Row2', 3, 4]])
df1 = pd.DataFrame(data=data[1:, 1:], index=data[1:, 0], columns=data[0, 1:])
print(df1.to_string())

# 2D array
my_2darray = np.array([[1, 2, 3], [4, 5, 6]]) 
df2 = pd.DataFrame(my_2darray)
print(df2.to_string())

# Dictionary to DataFrame
my_dict = {1: ['1', '3'], 2: ['1', '2'], 3: ['2', '4']} 
df3 = pd.DataFrame(my_dict)
print(df3.to_string())

# Column DataFrame
my_df = pd.DataFrame(data=[4, 5, 6, 7], index=range(0, 4), columns=['A']) 
print(my_df.to_string())

# Series to DataFrame
my_series = pd.Series({
    "UnitedKingdom": "London", 
    "India": "NewDelhi", 
    "United States": "Washington", 
    "Belgium": "Brussels"
}) 
df_series = pd.DataFrame(my_series, columns=['Capital'])
print(df_series.to_string())

# Shape and length
df = pd.DataFrame(np.array([[1, 2, 3], [4, 5, 6]])) 
print(df.shape) 
print(len(df.index))
```
